**Importing** **Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score ,recall_score, f1_score
from sklearn.model_selection import train_test_split
```

0 ---> Real Transaction.  

1 ---> Fradulant Transaction .

```python
# Seperating Values for Data Analysis
real = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
fraud.Amount.describe()
```

**Undersampling:**

The process where we decrease the size of majority class

```python
legit_sample = real.sample(n=492) # as there are 492 fradulant rows
```

```python
# Concatinating both the DataFrames
updated_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
updated_dataset.shape
```

```python
updated_dataset.head()
```

```python
updated_dataset.tail()
```

```python
# Shuffling the values here
df_shuffled = updated_dataset.sample(frac=1).reset_index(drop=True)
```

```python
df_shuffled.head()
```

```python
df_shuffled.tail()
```

```python
df_shuffled.value_counts('Class')
```

```python
df_shuffled.groupby('Class').mean()
```

Splitting the Data:

```python
X  = df_shuffled.drop(columns = 'Class', axis = 1)
Y = df_shuffled['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
Y.unique()
```

Splitting the Data into Training and Testing:

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

**Model Training:**

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score = accuracy_score(Y_test, y_pred)
recall = recall_score(Y_test, y_pred, average='macro')   # use 'macro' for multiclass
f1 = f1_score(Y_test, y_pred, average='macro')
```

```python
print("Accuracy Score:", accuracy_score)
print("Recall Score:", recall)
print("F1 Score:", f1)
```
